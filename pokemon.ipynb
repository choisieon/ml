```python
import pandas as pd

df = pd.read_csv('data/pokemon.csv')
```

```python
df.head(100)
```

```python
fire_df = df.loc[df['Type1'] == 'Fire']
```

```python
fire_height = fire_df[['Height (m)']]
fire_weight = fire_df[['Weight (kg)']]
#여기서 원본데이터의 필요없는 값을 제거해야 한다.
```

```python
fire_df.sort_values(by= 'Weight (kg)', ascending=False).head()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = \
train_test_split(fire_height, fire_weight)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
import matplotlib.pyplot as plt
plt.scatter(train_input, train_target)
plt.plot([0, 2.5], [0*60 -23, 2.5*60-23])
```

```python
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```
